# Process the raw data
## 1. Merge all the data

```python
import pandas as pd
```

```python

data_list = []

# read all the data
for year in range(1953, 2023):
    filename = f"{year}.csv"
    data = pd.read_csv(filename)
    
    # remove the columns with no names
    data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
    
    # append the data to data list
    data_list.append(data)

# merge all the tada
all_data = pd.concat(data_list, ignore_index=True)
```

```python
# check the general information of the data
all_data.info()
```

```python
# check the statistics of the data
all_data.describe()
```

```python
all_data.to_csv('all_data.csv', index=False)
```

```python
# select the coloum we need
lego_merged_data = all_data[['Number', 'Theme', 'Subtheme', 'Year', 'Set name', 'Pieces', 'RRP (GBP)', 'RRP (USD)', 'RRP (CAD)', 'RRP (EUR)', 'Value new (USD)', 'Value used (USD)', 'Launch date', 'Exit date']]
```

```python
lego_merged_data
```

```python
lego_merged_data.to_csv('lego_merged_data.csv', index=False)
```

```python
# check the general information of lego_data
lego_merged_data.info()
```

## 2. Create a new dataset of the number of set per them over  years
## (prepare for the analysis of theme trends)

```python
# calculate the set number of each theme per year
set_counts_per_year = lego_data.groupby(['Year', 'Theme']).size().reset_index(name='Set Count')
```

```python
set_counts_per_year.info()
```

```python
set_counts_per_year.to_csv('set_counts_per_year.csv', index=False)
```

## 3. Deal with the missing value of LEGO data to get a complete price data 

```python
columns_to_check = ['Value new (USD)', 'Value used (USD)', 'Launch date', 'Exit date', 'RRP (USD)']
lego_data_price = lego_data.dropna(subset = columns_to_check)
lego_data_price = lego_data_price.reset_index(drop=True)
lego_data_price = lego_data_price[['Number', 'Theme', 'Year', 'Set name', 'RRP (USD)', 'Value new (USD)', 'Value used (USD)', 'Launch date', 'Exit date']]
```

```python
lego_data_price.to_csv('lego_price_data.csv', index = False)
```
