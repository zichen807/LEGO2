```python
# filter the year from 2000 to 2022
filtered_data = count_data[(count_data['Year'] >= 2000) & (count_data['Year'] <= 2022)]
```

```python
# filter the the top 10 theme per year
top_themes_per_year = filtered_data.groupby('Year').apply(lambda x: x.nlargest(10, 'Set Count')).reset_index(drop=True)
```

```python
# check the unique theme
unique_themes = top_themes_per_year['Theme'].unique()
```

```python
# convert the data to pivot table
heatmap_data = top_themes_per_year.pivot("Theme", "Year", "Set Count")
```

```python
heatmap_data
```

```python
plt.figure(figsize=(20, 20))
ax = sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="YlGnBu", annot_kws={"size": 14}, cbar_kws={'shrink': 0.6})
plt.title('LEGO Set Count by Theme\n2000-2022', fontsize=20)
plt.ylabel('Theme', fontsize=18)
plt.xlabel('Year', fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
cbar= ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=14)
cbar.ax.set_title('Set count', fontsize=18)

plt.show()
```
